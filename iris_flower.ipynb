```python
x=df.iloc[:, :-1].values
y=df.iloc[:, -1].values
print(x)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y = y.ravel()
```

```python
for i, label in enumerate(le.classes_):
    print(f"{label}: {i}")
```

```python
plt.figure(figsize=(8,8))
sc=plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'],c= y, edgecolors='k',cmap='viridis')
cbar=plt.colorbar(sc)
cbar.set_ticks([0,1,2])
cbar.set_ticklabels(['Iris-setosa','Iris-versicolor','Iris-virginica'])
cbar.set_label('Species')
plt.xlabel('Petal Length Cm')
plt.ylabel("Petal Witdh Cm")
plt.title("Petal length by Species")

plt.show()
```

```python
plt.figure(figsize=(8,8))
sc=plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'],c= y, edgecolors='k',cmap='viridis')
cbar=plt.colorbar(sc)
cbar.set_ticks([0,1,2])
cbar.set_ticklabels(['Iris-setosa','Iris-versicolor','Iris-virginica'])
cbar.set_label('Species')
plt.xlabel('Petal Length Cm')
plt.ylabel("Petal Witdh Cm")
plt.title("Petal length by Species")

plt.show()
```

```python
plt.boxplot(x, tick_labels=['SepalLengthCm','SepalWitdhCm','PetalLengthcm','PetalWidthCm'],vert=True)
plt.show()
```

```python
species = df['Species'].unique()
x_species = [df[df['Species'] == s].drop(columns=['Species']).values for s in species]
```

```python
setosa_means = x_species[le.transform(['Iris-setosa'])[0]].mean(axis=0)
versicolor_means = x_species[le.transform(['Iris-versicolor'])[0]].mean(axis=0)
virginica_means = x_species[le.transform(['Iris-virginica'])[0]].mean(axis=0)

```

```python
cmap=plt.cm.magma
color=[cmap(0.2),cmap(0.5),cmap(0.8)]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, feature in enumerate(feature_names):
    ax = axes[i // 2, i % 2]
    means = [setosa_means[i], versicolor_means[i], virginica_means[i]]
    ax.bar([0, 1, 2], means,color=color, edgecolor=color, linewidth=2)
    ax.set_xticks([0, 1, 2])
    ax.set_xticklabels(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
    ax.set_title(feature)

fig.suptitle('Average Feature Values per Species', size=18)
fig.tight_layout()
fig.subplots_adjust(top=0.85)

plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=3,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
 from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
